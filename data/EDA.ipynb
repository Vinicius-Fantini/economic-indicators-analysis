```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.makedirs('../imgs', exist_ok=True)
```

```python
df_brasil = df[df['country_name'] == 'Brazil']

plt.figure(figsize=(10, 5))
plt.plot(df_brasil['year'], df_brasil['GDP (Current USD)'], marker='o')
plt.title('Evolução do PIB do Brasil')
plt.xlabel('Ano')
plt.ylabel('PIB')
plt.grid(True)
plt.savefig('../imgs/pib_brasil.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
paises = ['Brazil', 'India', 'China', 'United States']

df_paises = df[df['country_name'].isin(paises)]
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_paises, x='year', y='Inflation (CPI %)', hue='country_name', marker='o')
plt.title('Inflação – Brasil x Índia x China x EUA')
plt.ylabel('Inflação (%)')
plt.grid(True)
plt.savefig('../imgs/inflacao_brics_usa.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
mercosul = ['Brazil', 'Argentina', 'Paraguay', 'Uruguay']

df_mercosul = df[df['country_name'].isin(mercosul)]
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_mercosul, x='year', y='Inflation (CPI %)', hue='country_name', marker='o')
plt.title('Inflação Mercosul')
plt.ylabel('Inflação (%)')
plt.grid(True)
plt.savefig('../imgs/inflacao_mercosul.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
maiores_economias = ['United States', 'China']
df_eua_china = df[df['country_name'].isin(maiores_economias)]

plt.figure(figsize=(10, 5))
sns.lineplot(data=df_eua_china, x ='year', y = 'GDP (Current USD)', hue='country_name', marker='o')
plt.title('Evolução do PIB do EUA x China')
plt.xlabel('Ano')
plt.ylabel('PIB')
plt.grid(True)
plt.savefig('../imgs/pib_eua_china.png', dpi=300, bbox_inches='tight')
plt.show()
```
